```python
import face_recognition
import cv2
import numpy as np
```

```python


face_locations = []
face_encodings = []
face_names = []
process_this_frame = True

video_capture = cv2.VideoCapture(0)
for i in range(1,9):
    kanye_img = face_recognition.load_image_file(f"./pictures/008.jpg")

kanye_img_encoding = face_recognition.face_encodings(kanye_img)[0]

cna = face_recognition.load_image_file("./cna.jpeg")
cna_encoding = face_recognition.face_encodings(cna)[0]

known_face_encodings = [  kanye_img_encoding, cna_encoding ]

known_face_names = [ "kanye","cna"]

while True:
    
    ret, frame = video_capture.read() # single frame 
    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25) # make it small to speed up the process
    rgb_small_frame = small_frame[:, :, ::-1] # convert to rgb (face recognition needs rgb) 
    if process_this_frame:
        

        face_locations = face_recognition.face_locations(rgb_small_frame) # find all faces in the frame
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations) # encode all faces in the frame

        face_names = []
        for face_encoding in face_encodings:
            
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding) # compare all faces in the frame with the known faces
            name = "idk this face"

            if True in matches:
                first_match_index = matches.index(True) # find the index of the first match
                name = known_face_names[first_match_index] # get the name of the first match

            # Or instead, use the known face with the smallest distance to the new face
            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]: # if the best match is a match
                name = known_face_names[best_match_index] # get the name of the best match

            face_names.append(name) # append the name to the list of names

    process_this_frame = not process_this_frame


    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        
        top *= 4 # 0.25 * 4 = 1
        right *= 4
        bottom *= 4
        left *= 4

        # blur instead of face recognition 
        # face_image = frame[top:bottom, left:right]
        # face_image = cv2.GaussianBlur(face_image, (99, 99), 30)
        # frame[top:bottom, left:right] = face_image

        # Put the blurred face region back into the frame image
        

        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (10, 255, 111), 2)


        cv2.rectangle(frame, (left, bottom), (right, bottom), (10, 255, 111), cv2.FILLED)

        
        cv2.putText(frame, name, (left + 6, bottom - 6), cv2.FONT_HERSHEY_DUPLEX, 1.0, (255, 255, 255), 1)

        

    
    cv2.imshow('Face recognition test', frame)

    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()

# also can use a classifier to recognize faces
# https://github.com/ageitgey/face_recognition/blob/master/examples/face_recognition_svm.py
```
